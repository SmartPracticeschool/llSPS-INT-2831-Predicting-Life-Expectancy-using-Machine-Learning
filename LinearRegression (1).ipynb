```python
import numpy as np
import matplotlib.pyplot as pplot
```

```python
def get_regressor_bias(x, y):
    n = np.size(x)
    mean_x, mean_y = np.mean(x), np.mean(y)
    ss_xy = np.sum(x * y) - n * mean_x * mean_y
    ss_xx = np.sum(x * x) - n * mean_x * mean_x
    b1 = ss_xy / ss_xx
    b0 = mean_y - b1 * mean_x
    return (b0, b1)
```

```python
def plotter(x, y, b):
    pplot.scatter(x, y, color='m', marker='o', s=30)
    y_pred = b[0] + b[1] * x
    pplot.plot(x, y_pred, color='r')
    pplot.plot(x, y, color='b')
    pplot.xlabel("X")
    pplot.ylabel("Y")
    pplot.show()
```

```python
x = np.array([1, 2, 3, 5, 6, 7, 8, 8, 9])
y = np.array([1, 3, 3, 5, 6, 7, 7, 9, 9])
```

```python
b = get_regressor_bias(x, y)
print('The value of b0:{} and b1:{}'.format(b[0], b[1]))
plotter(x, y, b)
```
